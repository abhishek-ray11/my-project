```python
# Complete A/B test script for Jupyter (with UTF-8-safe file writing and preview)
# Requirements: pandas, numpy, scipy, matplotlib
# Usage: update file_A / file_B (or place your files in the notebook folder) and run cell.

import os
import math
import json
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# ------------------- USER CONFIG -------------------
# Set these to the filenames you've uploaded (or leave to auto-detect)
file_A = "final1"   # control
file_B = "final2"                    # treatment

# If you know the exact column and success value, set them here; otherwise leave None for auto-detection
col_to_use = None            # e.g. "Which option stood out to you most?"
success_value = None         # e.g. "Home button"

# Analysis options
bootstrap_iters = 10000
alpha = 0.05
output_report = "ab_test_report.txt"
output_plot = "ab_rates_plot.png"
# ----------------------------------------------------

def load_table(path):
    if not os.path.exists(path):
        return None
    ext = os.path.splitext(path)[1].lower()
    if ext in [".csv", ".txt"]:
        return pd.read_csv(path)
    elif ext in [".xls", ".xlsx"]:
        return pd.read_excel(path, sheet_name=0)
    else:
        try:
            return pd.read_csv(path)
        except:
            return pd.read_excel(path, sheet_name=0)

def autofind_files():
    # Attempt to use the configured file paths, else try to find likely CSV/XLSX files in cwd
    if os.path.exists(file_A) and os.path.exists(file_B):
        return file_A, file_B
    files = [f for f in os.listdir(".") if f.lower().endswith((".csv",".xlsx",".xls"))]
    if len(files) >= 2:
        return files[0], files[1]
    raise FileNotFoundError("Could not find two data files. Place them in the notebook folder and set file_A/file_B.")

def detect_candidate_columns(df):
    tokens_interest = {'yes','no','clicked','not clicked','click','home','review','search','a','b','home button'}
    candidates = []
    for col in df.columns:
        ser = df[col].dropna()
        if ser.empty: continue
        n_unique = ser.nunique(dropna=True)
        sample_tokens = set([str(x).strip().lower() for x in ser.unique()[:200]])
        if sample_tokens & tokens_interest:
            candidates.append(col)
        elif 2 <= n_unique <= 6:
            candidates.append(col)
    return candidates

def show_head_and_counts(df, name, n=6):
    print(f"\n--- Preview: {name} (shape {df.shape}) ---")
    display(df.head(n))
    print("\nColumn summaries:")
    for col in df.columns:
        ser = df[col].dropna()
        if ser.nunique() <= 12:
            print(f"  - {col}: {ser.nunique()} unique -> top values:")
            print(ser.astype(str).value_counts().head(10).to_string())
        else:
            print(f"  - {col}: {ser.nunique()} unique")

def map_to_binary(series, success_val=None):
    ser = series.dropna().astype(str).str.strip()
    if success_val is not None:
        return (ser == str(success_val)).astype(int)
    # numeric 0/1
    try:
        num = pd.to_numeric(ser, errors='coerce')
        uniq = set(num.dropna().unique())
        if uniq.issubset({0,1}):
            return num.fillna(0).astype(int)
    except:
        pass
    mapping = {'yes':1,'y':1,'true':1,'t':1,'1':1,'no':0,'n':0,'false':0,'f':0,'0':0,'clicked':1,'click':1,'not clicked':0,'home button':1}
    mapped = ser.str.lower().map(mapping)
    if mapped.notna().sum() / len(mapped) >= 0.5:
        return mapped.fillna(0).astype(int)
    return None

def two_prop_z_test(x1, n1, x2, n2):
    p1 = x1 / n1
    p2 = x2 / n2
    p_pool = (x1 + x2) / (n1 + n2)
    se_pool = math.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
    z = (p2 - p1) / se_pool if se_pool > 0 else float('nan')
    p_two = 2 * (1 - stats.norm.cdf(abs(z)))
    return {"p1":p1, "p2":p2, "p_pool":p_pool, "z":z, "p_two":p_two, "diff": p2 - p1}

def unpooled_ci(p1, n1, p2, n2, alpha=0.05):
    se = math.sqrt(p1*(1-p1)/n1 + p2*(1-p2)/n2)
    z = stats.norm.ppf(1 - alpha/2)
    diff = p2 - p1
    return diff - z*se, diff + z*se, se

def bootstrap_diff(a_vals, b_vals, iters=10000, seed=123):
    rng = np.random.RandomState(seed)
    diffs = []
    for _ in range(iters):
        sa = rng.choice(a_vals, size=len(a_vals), replace=True)
        sb = rng.choice(b_vals, size=len(b_vals), replace=True)
        diffs.append(np.mean(sb) - np.mean(sa))
    diffs = np.array(diffs)
    return np.percentile(diffs, [2.5, 97.5]), diffs.std(ddof=1), diffs

# ---------------- Main flow ----------------
A_path, B_path = autofind_files()
print("Using files:\n A:", A_path, "\n B:", B_path)

dfA = load_table(A_path)
dfB = load_table(B_path)

# show quick previews
show_head_and_counts(dfA, "Variant A")
show_head_and_counts(dfB, "Variant B")

# choose column
if col_to_use is None:
    candA = detect_candidate_columns(dfA)
    candB = detect_candidate_columns(dfB)
    print("\nAuto-candidates (A):", candA)
    print("Auto-candidates (B):", candB)
    common = [c for c in candA if c in candB]
    if common:
        col = common[0]
    else:
        # prefer exact same column name present in both
        same = [c for c in dfA.columns if c in dfB.columns]
        col = None
        for c in same:
            if c in candA or c in candB:
                col = c; break
        if col is None:
            col = candA[0] if candA else (candB[0] if candB else None)
    if col is None:
        raise SystemExit("Could not auto-detect a column. Set col_to_use to the exact column name and re-run.")
    print(f"\nAuto-chosen column: '{col}'")
else:
    col = col_to_use
    print(f"\nUser-specified column: '{col}'")

# choose success_value
if success_value is None:
    valsA = dfA[col].astype(str).str.strip().value_counts()
    valsB = dfB[col].astype(str).str.strip().value_counts()
    candidate = None
    for v in list(valsB.index) + list(valsA.index):
        if "home" in v.lower() or "button" in v.lower():
            candidate = v; break
    if candidate is None:
        candidate = valsB.idxmax() if len(valsB)>0 else (valsA.idxmax() if len(valsA)>0 else None)
    success_value = candidate
print("Auto-chosen success_value:", repr(success_value))

# map to binary
binA = map_to_binary(dfA[col], success_val=success_value)
binB = map_to_binary(dfB[col], success_val=success_value)

if binA is None or binB is None:
    print("\nAutomatic mapping failed. Sample values (A):")
    print(dfA[col].astype(str).value_counts().head(40).to_string())
    print("\nSample values (B):")
    print(dfB[col].astype(str).value_counts().head(40).to_string())
    raise SystemExit("Set success_value exactly (string) for which value should be treated as success and re-run.")

# compute stats
nA = len(binA); nB = len(binB)
clicksA = int(binA.sum()); clicksB = int(binB.sum())
pA = clicksA / nA; pB = clicksB / nB; diff = pB - pA

res = two_prop_z_test(clicksA, nA, clicksB, nB)
ci_low, ci_high, se_unp = unpooled_ci(pA, nA, pB, nB, alpha=alpha)
(boot_low, boot_high), boot_se, diffs = bootstrap_diff(binA.values, binB.values, iters=bootstrap_iters)

# save report (UTF-8 safe)
report_lines = [
    "A/B Test Analysis Report (automatically generated)",
    "",
    f"Files used: A: {A_path} (n={nA}), B: {B_path} (n={nB})",
    f"Column used: '{col}'",
    f"Success value: '{success_value}'",
    "",
    f"Variant A: clicks={clicksA}, n={nA}, click_rate={pA:.4f}",
    f"Variant B: clicks={clicksB}, n={nB}, click_rate={pB:.4f}",
    f"Observed difference (B - A) = {diff:.4f} ({diff*100:.2f} percentage points)",
    "",
    "Two-proportion z-test (two-sided):",
    f" - pooled proportion = {res['p_pool']:.6f}",
    f" - z-statistic = {res['z']:.6f}",
    f" - p-value (two-sided) = {res['p_two']:.6f}",
    "",
    f"95% CI (unpooled) = [{ci_low:.6f}, {ci_high:.6f}]",
    f"Bootstrap {bootstrap_iters}-iter CI = [{boot_low:.6f}, {boot_high:.6f}], bootstrap SE ≈ {boot_se:.6f}",
    "",
    "Interpretation: if p < alpha and CI excludes 0 => reject null (evidence of a difference).",
]

with open(output_report, "w", encoding="utf-8", errors="replace") as f:
    f.write("\n".join(report_lines))

print("\nReport saved to:", output_report)

# summary display
print("\n--- Summary ---")
print(f"Variant A: n={nA}, clicks={clicksA}, rate={pA:.4f}")
print(f"Variant B: n={nB}, clicks={clicksB}, rate={pB:.4f}")
print(f"Observed difference (B-A) = {diff:.4f} ({diff*100:.2f} pp)")
print(f"z = {res['z']:.4f}, p-value(two-sided) = {res['p_two']:.6f}")
print(f"95% CI (unpooled) = [{ci_low:.4f}, {ci_high:.4f}]")
print(f"Bootstrap 95% CI = [{boot_low:.4f}, {boot_high:.4f}] (SE ≈ {boot_se:.6f})")

summary_df = pd.DataFrame({
    "variant": ["A","B"],
    "file": [A_path, B_path],
    "n": [nA, nB],
    "success_count": [clicksA, clicksB],
    "success_rate": [pA, pB]
})
display(summary_df)

# plot and save
zcrit = stats.norm.ppf(1 - alpha/2)
ses = [math.sqrt(pA*(1-pA)/nA), math.sqrt(pB*(1-pB)/nB)]
cis = [zcrit*s for s in ses]

plt.figure(figsize=(6,4))
x = [0,1]
rates = [pA, pB]
plt.errorbar(x, rates, yerr=cis, fmt='o', capsize=6)
plt.xticks(x, ["A","B"])
plt.ylim(0, min(1, max(rates) + 0.15))
plt.ylabel("Success rate")
plt.title(f"Success rate for '{success_value}' with {int((1-alpha)*100)}% CI")
plt.grid(True)
plt.tight_layout()
plt.savefig(output_plot)
plt.show()
print("Plot saved to:", output_plot)

# save value counts for inclusion in report
vcA = dfA[col].astype(str).str.strip().value_counts().to_dict()
vcB = dfB[col].astype(str).str.strip().value_counts().to_dict()
with open("value_counts_question.json", "w", encoding="utf-8") as f:
    json.dump({"A": vcA, "B": vcB}, f, indent=2, ensure_ascii=False)

print("\nSaved value counts to value_counts_question.json")
print("\nDone. If you want the script to use a different column or success_value, set col_to_use or success_value and re-run.")
```
